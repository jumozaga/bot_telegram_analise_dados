```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv("../data/raw/avaliacao_mensal_funcionarios.csv")
file['NPS interno'] = file['NPS interno'].str.replace(",", ".").astype("float")
file
```

```python
file.Setor = file.Setor.replace({"Engenheiro de Software":"Engenharia de Software"})
```

```python
# NPS INTERNO MENSAL MÉDIO POR SETOR

setores = file.groupby(["Setor"]).mean()["NPS interno"].index
nps_medio_setor = file.groupby(["Setor"]).mean()["NPS interno"].values
plt.bar(setores, nps_medio_setor);
# plt.xlabel("SETORES")
plt.ylabel("NPS interno");
plt.xticks(rotation= 55);
plt.yticks(np.arange(0, 11, 1));
plt.title("Média de NPS Interno Mensal Por Setor");
```

```python
# NPS INTERNO MENSAL MÉDIO POR CONTRATAÇÃO

tipo_contratacao = file.groupby(["Tipo de Contratação"]).mean()["NPS interno"].index
nps_medio_tipo_contratacao = file.groupby(["Tipo de Contratação"]).mean()["NPS interno"].values
plt.bar(tipo_contratacao, nps_medio_tipo_contratacao);
# plt.xlabel("SETORES")
plt.ylabel("NPS interno");
plt.xticks(rotation= 55);
plt.yticks(np.arange(0, 11, 1));
plt.title("Média de NPS Interno Mensal Por Tipo de Contratação");
```

```python
# DISTRIBUIÇÃO DO NPS INTERNO
# Histograma
plt.hist(file["NPS interno"])
plt.title("Distribuição do NPS Interno mensal")
plt.xlabel("NPS Interno");

```
